# Plot observed data, overlap with expected signal and background (normalized to observed yields) with optimal event selection.

```python
origArr1 = len(qcd['mass'])
origSig1 = len(qcd['mass'])

word = 'mass'
newArr1Mass = []
newSig1Mass = []
arr1Ind = []
sig1Ind = []
for i in range(len(qcd[word])):
    if(qcd[word][i] < 130 and qcd[word][i] > 120):
        newArr1Mass.append(qcd[word][i])
        arr1Ind.append(i)
    if(higgs[word][i] < 130 and higgs[word][i] > 120):
        newSig1Mass.append(higgs[word][i])
        sig1Ind.append(i)
print('Significance (Mass cuts 120-130): ', (100 * (len(newSig1Mass) / origSig1)) / (20000 * (len(newArr1Mass) / origArr1))**.5)
plt.hist(newSig1Mass, 500, density=True)
plt.hist(newArr1Mass, 500, density=True, color='red', alpha=.5)
plt.title('Stacked Distribution with Mass Cuts from 120-130')
plt.xlabel('3-subjettiness')
plt.ylabel('Probability')
plt.show()
```

- Observed Data: 2.42 -> 1.71 -> 1.99

Our signficance seems to decrease drastically as we replace our QCD with Observed high luminosity data in our significance calculation. This makes sense mathemtatically because high luminosity is like a high signal source of something such as a star. With high luminosity you have a higher number of data point and so in the equation signif. = NHiggs / sqrt(NObs) as NObs increases significance decreases.

For the last plot we first applied mass cuts to both D2 and then T2 and then made further mass cuts on first D2 and then T2. Using two features to make optimizing cuts improves the significance a lot more and we can see even in this case the high luminosity data makes the signficance decrease when taking cuts at the same sections.

# Evalute the 1   uncertainty of the expected 95% confidence level upper limit.

```python
oneSig = stats.norm.cdf(1)
sigLoc = stats.poisson.ppf(oneSig, sig)
print("Signal has 1sigma uncertainty: ", abs(sig - sigLoc))
backLoc = stats.poisson.ppf(oneSig, back)
print("Background has 1sigma uncertainty: ", abs(back - backLoc))
```

To do this calculation we took the we found the probability of one sigma on a normal distribution to get the standard probability of 1 sigma. Then we found where on a poisson distribution, with a mean of the 95% confidence interval of the signal, matched one sigma probability. This would give us some number and we subtract that by the 95% confidence interval we calculated to find the true one sigma. We do this same process for background.

We find that the expected confidence is a good approximation of signal because the expected values 1 sigma 95% confidence intervals are well within one sigma of one another.
